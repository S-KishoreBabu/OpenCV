```python
import cv2
from matplotlib import pyplot as plt
file_path = 'Assets/sample.jpg'

img = cv2.imread(file_path)

height,width,color = img.shape
w = width//2
h = height//2
reSized_img = cv2.resize(img,(w,h))
plt.imshow(cv2.cvtColor(reSized_img,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
import numpy as np

img = cv2.imread(file_path)
flip = cv2.flip(img,0)
resizeFlip = cv2.resize(flip,(w,h))
plt.imshow(cv2.cvtColor(resizeFlip,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
flip = cv2.flip(img,1)
resizeFlip = cv2.resize(flip,(w,h))
plt.imshow(cv2.cvtColor(resizeFlip,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
flip = cv2.flip(img,-1)
resizeFlip = cv2.resize(flip,(w,h))
plt.imshow(cv2.cvtColor(resizeFlip,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
bwimage = cv2.imread(file_path,0)
plt.imshow(cv2.cvtColor(bwimage,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
return_value,binary_threshold = cv2.threshold(bwimage,130,255,cv2.THRESH_BINARY)
plt.imshow(cv2.cvtColor(binary_threshold,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
colorImage = cv2.imread(file_path,1)
return_value,binary_threshold = cv2.threshold(colorImage,150,255,cv2.THRESH_BINARY)
plt.imshow(cv2.cvtColor(binary_threshold,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
blured_img = cv2.GaussianBlur(colorImage,(199,199),0)
plt.imshow(cv2.cvtColor(blured_img,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
